```python
# import the function
from urllib.request import urlopen
```

```python
url = urlopen('https://books.toscrape.com/')
```

```python
text = url.read()
```

```python
print(text.decode('utf-8'))
```

```python
from nltk.tokenize import word_tokenize
```

```python
pip install bs4
```

```python
# import the class
from bs4 import BeautifulSoup
```

```python
# create the soup object
soup = BeautifulSoup(text, 'html.parser')
```

```python
data = soup.get_text()
```

```python
print(data)
```

```python
tokens1 = word_tokenize(data)
```

```python
tokens1
```

```python
# import the class
from nltk import FreqDist
```

```python
# create the object 
freq = FreqDist(tokens1)
```

```python
freq.plot(20)
```

```python
# Remove the punctuation
tokens2 = [x for x in tokens1 if x.isalpha()]
```

```python
freq = FreqDist(tokens2)
freq.plot(20)
```

```python
# remove the stopwords
from nltk.corpus import stopwords
swords = stopwords.words('english')
```

```python
tokens3 = [x.lower() for x in tokens2 if x.lower() not in swords]
```

```python
freq = FreqDist(tokens3)
freq.plot(20)
```

```python
# stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
tokens4 = [ps.stem(x) for x in tokens3]
```

```python
freq = FreqDist(tokens4)
freq.plot(20)
```

```python
freq.most_common(10)
```
